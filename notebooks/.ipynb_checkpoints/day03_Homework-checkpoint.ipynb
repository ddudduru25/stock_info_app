# 잠깐! 매개변수와 인자

**매개변수(parameter)**는 *함수의 입력값을 함수 내부로 전달하는 데 사용하는 변수*로, 함수를 정의할 때 사용해요. **인수(argument)**는 *함수로 전달하는 입력값으로, 함수를 호출할 때 사용*해요. 매개변수와 인수는 둘 다 함수에서 사용하는 용어로 구분해 사용하기도 하지만 보통 혼용하기도 합니다. 참고로 매개변수는 **인자**라고도 합니다.

# [실습 4] 3.1.1 함수의 정의와 호출 - 매개변수 2개, 반환값 O  ___ 84쪽

성, 이름과 함께 안녕하세요를 문자열로 반환하는 *get_hello()*라는 함수를 만드세요. 함수에서 반환할 때는 return을 활용해야 해요.

*get_hello()*를 호출하여 "윤 히히 님! 안녕하세요"를 hello_text 변수에 저장하세요.

값이 잘 저장됐는지 확인하기 위해 hello_text를 출력해보세요. [실습 3]과 달리 이번 get_hello 함수는 호출해도 아무런 일도 일어나지 않는 것처럼 보여요. 반환받은 값을 변수에 저장하고, 그 변수를 print() 함수로 출력해야만 결과를 볼 수 있답니다. 

``` python
# 힌트

# 함수 정의하기(만들기)
def 함수명(성, 이름):
  return "X OO 님! 안녕하세요"  # return으로 반환

# 함수 사용하기(호출하기)
변수명 = 함수명(성1, 이름1)       # 반환받은 값을 저장할 변수가 필요함

# 변수 출력하기
print(변수명)                  # 실습 3과 달리 print 함수를 활용해 출력해야 함
```

````python
# 여기에 코드를 작성해주세요.
def get_hello(Lname, name):
    return (Lname + name + "님! 안녕하세요")

hello_text = get_hello("김", "제니")

print(hello_text)
````

# 한 걸음 더

오늘 공부한 모듈, 함수는 최종 코드에서 다음과 같이 활용됩니다. 이해가 안 되는 게 당연해요! 오늘 배운 내용만 쪼개서 살펴볼게요. 그래도 def, return 등의 익숙한 코드가 보이죠?

---
``` python
import streamlit as st
import pandas as pd
import yfinance as yf
import datetime
import matplotlib.pyplot as plt
import matplotlib
from io import BytesIO

#----------------------------------------
# 한국 주식 종목 코드를 가져오는 함수
#----------------------------------------
def get_stock_info(market_type=None):
  # 한국거래소(KRX)에서 전체 상장법인 목록 가져오기
  base_url =  "http://kind.krx.co.kr/corpgeneral/corpList.do"
  method = "download"
  if market_type == 'kospi':
    marketType = "stockMkt"  # 주식 종목이 코스피인 경우
  elif market_type == 'kosdaq':
    marketType = "kosdaqMkt" # 주식 종목이 코스닥인 경우
  elif market_type == None:
    marketType = ""
  url = "{0}?method={1}&marketType={2}".format(base_url, method, marketType)

  df = pd.read_html(url, header=0)[0]
  
  # 종목코드 열을 6자리 숫자로 표시된 문자열로 변환
  df['종목코드']= df['종목코드'].apply(lambda x: f"{x:06d}")
  
  # 회사명과 종목코드 열 데이터만 남김
  df = df[['회사명','종목코드']]
  
  return df

#----------------------------------------------------
# yfinance에 이용할 Ticker 심볼을 반환하는 함수
#----------------------------------------------------
def get_ticker_symbol(company_name, market_type):
  df = get_stock_info(market_type)
  code = df[df['회사명']==company_name]['종목코드'].values
  code = code[0]
  
  if market_type == 'kospi':
    ticker_symbol = code +".KS" # 코스피 주식의 심볼
  elif market_type == 'kosdaq':
    ticker_symbol = code +".KQ" # 코스닥 주식의 심볼
  
  return ticker_symbol
#---------------------------------------------------------

# 주식 종목과 종류 지정해 ticker 심볼 획득
ticker_symbol = get_ticker_symbol(stock_name, "kospi")
```

## 2단계

get_stock_info() 함수는 한국 거래소(KRX)에서 코스피/코스닥 주식 정보 데이터를 가져오는 함수에요. 

- 매개변수 market_type으로는 주식 종목이 코스피인지 코스닥인지 전달받습니다.
- 최종 코드에서는 market_type에 따라 한국 거래소(KRX)에서 주식 정보 데이터를 가져옵니다.
- 가져온 주식 정보 데이터를 반환합니다.

````python
# ----------------------------------------
# 한국 주식 종목 코드를 가져오는 함수
# ----------------------------------------
def get_stock_info(market_type=None):
  if market_type == 'kospi':
    marketType = "stockMkt"  # 주식 종목이 코스피인 경우
  elif market_type == 'kosdaq':
    marketType = "kosdaqMkt" # 주식 종목이 코스닥인 경우
  elif market_type == None:
    marketType = ""
  
  df = marketType + " 데이터를 가져왔습니다."

  return df
````

````python
# ----------------------------------------
# 인자로 kospi를 전달하여 get_stock_info 호출해보기
# ----------------------------------------
df = get_stock_info("kospi")
print(df)
````

````python
# ----------------------------------------
# 인자로 kosdaq을 전달하여 get_stock_info 호출해보기
# ----------------------------------------
df = get_stock_info("kosdaq")
print(df)
````

## 3단계

get_ticker_symbol() 함수는 회사명에 해당하는 종목 코드를 가져오는 함수에요. 

- 매개변수 company_name, market_type을 전달받습니다. 
- company_name은 회사명을 의미하고, market_type은 주식 종목이 코스피인지 코스닥인지를 의미합니다.

- get_stock_info() 함수를 호출해 주식 정보 데이터를 가져옵니다.
- 가져온 주식 정보 데이터에서 company_name에 해당하는 종목 코드를 가져옵니다.
- market_type에 따라 티커 심볼을 지정합니다.

- 티거 심볼을 반환합니다.

````python
# ----------------------------------------------------
# yfinance에 이용할 Ticker 심볼을 반환하는 함수
# ----------------------------------------------------
def get_ticker_symbol(company_name, market_type):
  df = get_stock_info(market_type)
  print(df)

  code = "가져온 데이터(df)에서 " + company_name + "의 종목 코드를 가져왔습니다. "

  if market_type == 'kospi':
    ticker_symbol = code + "[코스피]" # 코스피 주식의 심볼
  elif market_type == 'kosdaq':
    ticker_symbol = code +"[코스닥]" # 코스닥 주식의 심볼

  return ticker_symbol
# ---------------------------------------------------------
````

````python
# ----------------------------------------
# 인자로 naver, kospi를 전달하여 get_ticker_symbol 호출해보기
# ----------------------------------------
symbol = get_ticker_symbol("naver", "kospi")
print(symbol)
````

````python
# ----------------------------------------
# 인자로 studiodragon, kospi를 전달하여 get_ticker_symbol 호출해보기
# ----------------------------------------
symbol = get_ticker_symbol("studiodragon", "kosdaq")
print(symbol)
````
