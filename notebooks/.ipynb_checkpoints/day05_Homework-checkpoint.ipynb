# [실습 4] 1번 주석 구현하기

마켓 타입에 맞춰 URL 주소를 구성해주세요.



한국 거래소(KRX)에서 *코스피* 상장법인 목록을 내려받을 수 있는 주소는 아래와 같습니다. marketType= 뒤의 주소가 stockMkt입니다.

- http://kind.krx.co.kr/corpgeneral/corpList.do?method=download&marketType=stockMkt

한국 거래소(KRX)에서 *코스닥* 상장법인 목록을 내려받을 수 있는 주소는 아래와 같습니다. marketType= 뒤의 주소가 kosdaqMkt입니다.

- http://kind.krx.co.kr/corpgeneral/corpList.do?method=download&marketType=kosdaqMkt

&nbsp;

1) 변수 만들기
- 주소에서 공통되는 앞 부분을 base_url로 지정합니다. 
- http://kind.krx.co.kr/corpgeneral/corpList.do

- method 변수를 만들고, 값을 download로 지정합니다.


2) 조건문을 작성하고, 매개변수로 전달 받은 market 타입에 따라 주소의 뒷부분을 구성해주세요.
- 매개변수 market_type이 kospi라면 marketType이 stockMkt
- 매개변수 market_type이 kosdaq라면 marketType이 kosdaqMkt
- 매개변수 market_type이 없다면(None) marketType도 빈 값("")으로 구성합니다.

3) 이제 주소를 조립할 차례에요.
- format() 함수를 이용해 주소를 조립합니다.
- [공통되는주소]?method=[method 변수의 값]&marketType=[marketType 변수의 값]

&nbsp;

변수, 조건문, format() 함수가 잘 기억나지 않는다면 **2일차 실습 노트**를 살펴보세요.

````python
def get_stock_info(market_type=None):
  # 1. 마켓 타입에 맞춰 URL 주소 구성하기
  base_url =  "http://kind.krx.co.kr/corpgeneral/corpList.do"
  method = "download"
  if market_type == 'kospi':
    marketType = "stockMkt"  # 주식 종목이 코스피인 경우
  elif market_type == 'kosdaq':
    marketType = "kosdaqMkt" # 주식 종목이 코스닥인 경우
  elif market_type == None:
    marketType = ""
  url = "{0}?method={1}&marketType={2}".format(base_url, method, marketType)

  # 2. 한국 거래소에서 주식 정보 데이터 가져오기

  # 3. 가져온 주식 정보 데이터에서 필요한 데이터만 남기고, 알맞게 가공하기

  # 4. 주식 정보 데이터 반환
  return df
````

# [실습 5] 2번 주석 구현하기

1에서 만든 URL을 토대로 한국 거래소에서 주식 정보 데이터를 가져오세요.

&nbsp;

판다스의 read_html() 함수를 이용합니다.

판다스는 웹 사이트의 표 형식의 데이터를 쉽게 가져올 수 있는 read_html()을 제공합니다. 다만 웹 사이트가 자바스크립트로 구성돼 있다면 read_html()을 이용해 표 데이터를 가져오지 못할 수도 있습니다. 판다스의 read_html()의 사용법은 다음과 같습니다.

``` python
import pandas as pd

dfs = pd.read_html(url 혹은 html [, header = row_num] [,options])
```


read_html() 함수의 자세한 설명은 **258쪽 하단 ~ 260쪽**을 살펴보세요.

&nbsp;

``` python
# 힌트
데이터프레임 = pd.read_html(URL 주소, 헤더여부)[인덱스]

# 엑셀 파일의 첫 번째 줄에 헤더가 있다면 header=0을 지정
# 결괏값이 하나라면 인덱스로 [0]을 지정
데이터프레임 = pd.read_html(URL 주소, header=0)[0]
```

````python
def get_stock_info(market_type=None):
  # 1. 마켓 타입에 맞춰 URL 주소 구성하기
  base_url =  "http://kind.krx.co.kr/corpgeneral/corpList.do"
  method = "download"
  if market_type == 'kospi':
    marketType = "stockMkt"  # 주식 종목이 코스피인 경우
  elif market_type == 'kosdaq':
    marketType = "kosdaqMkt" # 주식 종목이 코스닥인 경우
  elif market_type == None:
    marketType = ""
  url = "{0}?method={1}&marketType={2}".format(base_url, method, marketType)

  # 2. 한국 거래소에서 주식 정보 데이터 가져오기
  df = pd.read_html(url, header=0)[0]

  # 3. 가져온 주식 정보 데이터에서 필요한 데이터만 남기고, 알맞게 가공하기

  # 4. 주식 정보 데이터 반환
  return df
````

# [실습 6] 3번 주석 구현하기

2에서 가져온 주식 정보 데이터에서 필요한 데이터만 남기고, 알맞게 가공해주세요.
- 종목 코드를 살펴보면 4자리 숫자부터 6자리 숫자까지 다양하게 구성돼 있어요. 모두 6자리로 표현할 수 있게, 6자리가 아닌 종목코드는 앞에 0을 붙여주세요.
- 주식 정보 데이터는 회사명, 종목코드, 업종 등 총 9개의 열로 구성돼 있어요.
- 우리는 회사명과 종목코드만 있으면 되므로 두 개의 열만 남겨주세요.

``` python
# 힌트
df['종목코드']= df['종목코드'].apply(lambda x: f"{x:06d}")
df = df[['가져오고자 하는 열1', '가져오고자 하는 열2']]
```

&nbsp;

원하는 열만 남기는 방법이 잘 기억나지 않는다면 **4일차 실습노트**를 살펴보세요.

종목 코드를 6자리 숫자로 구성하는 방법은 **252쪽 하단**을 살펴보세요.

````python
def get_stock_info(market_type=None):
  # 1. 마켓 타입에 맞춰 URL 주소 구성하기
  base_url =  "http://kind.krx.co.kr/corpgeneral/corpList.do"
  method = "download"
  if market_type == 'kospi':
    marketType = "stockMkt"  # 주식 종목이 코스피인 경우
  elif market_type == 'kosdaq':
    marketType = "kosdaqMkt" # 주식 종목이 코스닥인 경우
  elif market_type == None:
    marketType = ""
  url = "{0}?method={1}&marketType={2}".format(base_url, method, marketType)

  # 2. 한국 거래소에서 주식 정보 데이터 가져오기
  df = pd.read_html(url, header=0)[0]

  # 3. 가져온 주식 정보 데이터에서 필요한 데이터만 남기고, 알맞게 가공하기
  # 종목코드 열을 6자리 숫자로 표시된 문자열로 변환
  df['종목코드']= df['종목코드'].apply(lambda x: f"{x:06d}")
    
  # 회사명과 종목코드 열 데이터만 남김
  df = df[['회사명','종목코드']]

  # 4. 주식 정보 데이터 반환
  return df
````
